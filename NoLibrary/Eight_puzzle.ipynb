```python
from copy import deepcopy
from time import time
```

```python
def compare_history(history, moveset):
    if moveset in history:
        return 0
    else:
        return 1
```

```python
class Puzzle():
    def __init__(self, Parent, state, action):
        
        self.parent = Parent
        if Parent != None:
            self.state = state
            self.path_cost = Parent.path_cost + 1
            self.depth = Parent.depth + 1
            self.actions = action
            
        else:
            self.state = state
            self.path_cost = 0
            self.depth = 0
            self.actions = []
            
#        self.algorithm = algorithm

    def find_zero(self):
        loc_z = [c for c, i in enumerate(self.state) if i == 0][0]
        if loc_z >= 0 and loc_z <= 8:
            return loc_z
        else:
            raise ValueError('Index is out of range')
    
    def check_actions(self):
        lateral_move_rest_right = [2,5,8]
        lateral_move_rest_left = [0,3,6]
        zero_loc = self.find_zero()
        possible_states = dict()
        
        if zero_loc > 2:
            temp_state = deepcopy(self.state)
            temp_state[zero_loc], temp_state[zero_loc - 3] = \
            temp_state[zero_loc - 3], temp_state[zero_loc]
            possible_states.update({"Up" : temp_state})
            
        if zero_loc < 6:
            temp_state = deepcopy(self.state)
            temp_state[zero_loc], temp_state[zero_loc + 3] = \
            temp_state[zero_loc + 3], temp_state[zero_loc] 
            possible_states.update({"Down" : temp_state})
            
        if zero_loc not in lateral_move_rest_right:
            temp_state = deepcopy(self.state)
            temp_state[zero_loc], temp_state[zero_loc + 1] = \
            temp_state[zero_loc + 1], temp_state[zero_loc] 
            possible_states.update({"Right" : temp_state})
            
        if zero_loc not in lateral_move_rest_left:
            temp_state = deepcopy(self.state)
            temp_state[zero_loc], temp_state[zero_loc - 1] = \
            temp_state[zero_loc - 1], temp_state[zero_loc] 
            possible_states.update({"Left" : temp_state})
            
        if len(possible_states) < 0:
            raise ValueError("No movement available")
        else:
            return possible_states
        
    def action_dfs(self, history, stack):
        stack.pop(0)
        if self.score_state() == 9:
            print(f"Final State >> {self.state}")
            print(f"Path_cost >> {self.path_cost}")
            print(f"Depth >> {self.depth}")
            print(f"Movesets >> {self.actions}")
            return True
        else:
            available_states = self.check_actions()
            for move, new_state in available_states.items():
                if compare_history(history, new_state):
                    history.append(new_state)
                    new_action = deepcopy(self.actions)
                    new_action.append(move)
                    child = Puzzle(self, new_state, new_action)
                    stack.insert(0, child)
            return stack, history
                     
    
    def action_bfs(self, queue, history):
        if self.score_state() == 9:
            print(f"Final State >> {self.state}")
            print(f"Path_cost >> {self.path_cost}")
            print(f"Depth >> {self.depth}")
            print(f"Movesets >> {self.actions}")
            return True
        
        else:
            available_states = self.check_actions()
            for move, new_state in available_states.items():
                if compare_history(history, new_state):
                    history.append(new_state)
                    new_action = deepcopy(self.actions)
                    new_action.append(move)
                    child = Puzzle(self, new_state, new_action)
                    queue.append(child)
            queue.pop(0)
            return queue, history
        
    def score_state(self):
        return sum([1 for c, y in enumerate(self.state) if y-c == 0])
```

```python
initial_state = [8, 1, 3, 4, 0, 2, 7, 6, 5]
history = [[8, 1, 3, 4, 0, 2, 7, 6, 5]]
queue = []
eight_puzzle = Puzzle(None, initial_state, None)
stack = []
stack.append(eight_puzzle)
```

```python
s_t = time()
queue, history = eight_puzzle.action_bfs(queue, history)
print("BFS Algorithm")
while True:
    if len(queue):
        q = queue[0]
        try :
            queue, history = q.action_bfs(queue, history)
        except TypeError as e:
            print("The solution has found.")
            break
    else:
        print("The queue is empty which means no possible solution available.")
        break
f_t = time()
print(f"Whole operation lasted for {(f_t - s_t):3f} seconds.")
```
